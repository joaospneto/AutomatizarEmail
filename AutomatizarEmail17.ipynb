```python
import pandas as pd
```

```python
vendas_df = pd.read_excel('/content/Registro - Vendas.xlsx')
faturamento_total = vendas_df['Ticket da Venda'].sum()
print(faturamento_total)
vendas_df
```

```python
lista_df = pd.read_excel('/content/Lista Emails.xlsx')
lista_df
```

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

# ConfiguraÃ§Ãµes do servidor SMTP do Gmail
smtp_server = "smtp.gmail.com"
smtp_port = 587
login = "joaote@gmail.com"
senha = "743884"

# ConfiguraÃ§Ã£o do email
for i, email in enumerate(lista_df['E-mail']):
    nome = lista_df.loc[i, 'Nome'].strip()

    # Criar a mensagem
    msg = MIMEMultipart()
    msg['From'] = login
    msg['To'] = email
    msg['Subject'] = "Faturamento do mÃªs de Agosto"

    # Corpo do email em HTML
    body = (
        f"Oi {nome},<br><br>"
        "Tudo bem? Estou te enviando um relatÃ³rio do mÃªs de Agosto!<br>"
        "Segue em anexo o registro das vendas!<br><br>"
        f"O faturamento do mÃªs foi {faturamento_total}."
    )
    msg.attach(MIMEText(body, 'html'))


    try:
        server = smtplib.SMTP(smtp_server, smtp_port)
        server.starttls()
        server.login(login, senha)
        server.send_message(msg)
        server.quit()
        print(f"Email enviado com sucesso para: {email}")
    except Exception as e:
        print(f"Erro ao enviar email para {email}: {str(e)}")
```
